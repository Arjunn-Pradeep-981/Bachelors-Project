```python
def initialize_grid(total_time=5.0, Nx=128, Ny=128, Lx=128.0, Ly=128.0, dt=0.01):

    # Parameters for the domain
    dx, dy = Lx / Nx, Ly / Ny  # Grid spacing
    steps = int(total_time / dt)

    # Create spatial grid centered at (0,0)
    x = np.linspace(-Lx / 2, Lx / 2, Nx, endpoint=False)
    y = np.linspace(-Ly / 2, Ly / 2, Ny, endpoint=False)
    X, Y = np.meshgrid(x, y)

    # Wavenumber grids in Fourier space (centered)
    kx_unshifted = 2 * np.pi * fftfreq(Nx, d=dx)
    ky_unshifted = 2 * np.pi * fftfreq(Ny, d=dy)
    kx = fftshift(kx_unshifted)
    ky = fftshift(ky_unshifted)
    KX, KY = np.meshgrid(kx, ky)  # Shift to center k-space around (0,0)

    K_squared = KX**2 + KY**2
    K = np.sqrt(K_squared)
    K_fourth = K_squared**2

    # Handling division by zero at k=0
    K[K == 0] = 1e-20
    center_idx = Nx // 2

    return Nx, Ny, Lx, Ly, dx, dy, dt, total_time, steps, x, y, X, Y, kx, ky, KX, KY, K_squared, K, K_fourth, center_idx

def fft_transform(field, normalize=True):
    transformed = fftshift(fft2(field))
    if normalize:
        transformed /= field.size  # Normalize by grid size
    return transformed

def ifft_transform(field_tilde, normalize=True, real_output=True):
    field = ifft2(ifftshift(field_tilde))
    if normalize:
        field *= field_tilde.size  # Normalize by grid size
    return np.real(field) if real_output else field


def plot_3d(x, y, z, title):
    
    fig = plt.figure(figsize=(6, 6))
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot the surface
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('w')
    ax.view_init(elev=45, azim=-135)
    ax.plot_surface(x, y, z, cmap='viridis', rstride=1, cstride=1, antialiased=True)

    mappable = plt.cm.ScalarMappable(cmap='viridis')
    mappable.set_array(z)
    plt.colorbar(mappable, ax=ax, shrink=0.5, aspect=10)
    plt.show()

def plot_2d(x, y, z, title):
    
    plt.figure(figsize=(6, 6))
    plt.pcolormesh(x, y, z, cmap='viridis')
    plt.colorbar(label='w')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.gca().set_aspect('equal', adjustable='box')  # Set aspect ratio to equal
    plt.show()

def animate_w_and_w_tilde(w_array, w_tilde_array, X, Y, KX, KY, dt, title):
    # Initialize the figure
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    # Setup for w (real space)
    real_space_plot = ax2.pcolormesh(X, Y, w_array[0], cmap='viridis', shading='auto')
    ax2.set_title('Real Space (w)')
    ax2.set_xlabel('x')
    ax2.set_ylabel('y')
    ax2.set_aspect('equal')
    real_colorbar = plt.colorbar(real_space_plot, ax=ax2)

    # Setup for w_tilde (Fourier space)
    fourier_space_plot = ax1.pcolormesh(KX, KY, np.abs(w_tilde_array[0]), cmap='viridis', shading='auto')
    ax1.set_title('Fourier Space (w_tilde)')
    ax1.set_xlabel('kx')
    ax1.set_ylabel('ky')
    ax1.set_aspect('equal')
    fourier_colorbar = plt.colorbar(fourier_space_plot, ax=ax1)

    # Update function for animation
    def update(frame):
        real_space_plot.set_array(w_array[frame].ravel())
        real_space_plot.set_clim(vmin=w_array[frame].min(), vmax=w_array[frame].max())  # Dynamically rescale
        real_colorbar.update_normal(real_space_plot)

        # Update Fourier space data
        fourier_space_plot.set_array(np.abs(w_tilde_array[frame]).ravel())
        fourier_space_plot.set_clim(vmin=np.abs(w_tilde_array[frame]).min(), vmax=np.abs(w_tilde_array[frame]).max())  # Dynamically rescale
        fourier_colorbar.update_normal(fourier_space_plot)

        # Update titles with the current time
        ax2.set_title(f'Real Space (w), Time: {frame * dt:.2f}')
        ax1.set_title(f'Fourier Space (w_tilde), Time: {frame * dt:.2f}')
        return fourier_space_plot, real_space_plot

    # Create the animation
    ani = FuncAnimation(fig, update, frames=len(w_array), blit=False)
    ani.event_source.interval = 2000  # Set the interval between frames in milliseconds
    # Save the animation as a video file or show it
    ani.save(title, fps=10, extra_args=['-vcodec', 'libx264'])
    plt.show()

def generate_smooth_random_2d_array(size=(100, 100), sigma=1.0):
    # Generate random noise
    noise = np.random.normal(size=size)
    
    # Apply Gaussian filter to smooth the noise
    smooth_array = gaussian_filter(noise, sigma=sigma)
    
    return smooth_array

# plt.imshow(smooth_array, cmap='viridis')
# plt.colorbar()
# plt.title('Smooth Random 2D Array (Gaussian Filter)')
# plt.show()

Nx, Ny, Lx, Ly, dx, dy, dt, total_time, steps, x, y, X, Y, kx, ky, KX, KY, K_squared, K, K_fourth, center_idx = initialize_grid()

# w = np.exp((-X*X - Y*Y)/4086) * 0.1

# w = np.sin(2 * np.pi * X / 32)

# w_tilde = fft_transform(w)

# plot_2d(X, Y, w, 'Initial condition')
# plot_2d(KX, KY, np.abs(w_tilde), 'Initial condition in Fourier space')

# plot_3d(X, Y, w, 'Initial condition')
# plot_3d(KX, KY, np.abs(w_tilde), 'Initial condition in Fourier space')
```

## Linear parts

```python
def linear_first_order_scheme_for_w(w_tilde, dt, K_fourth, K_squared, K, f_el_prefactor, gamma0):

    w_tilde_new = w_tilde / (1 + dt * (K_fourth - gamma0 * K_squared + 2*K*f_el_prefactor))

    return w_tilde_new

def linear_first_order_scheme_for_u1(u1_tilde, dt, K_squared, K, f_el_prefactor, KY):

    u1_tilde_new = -u1_tilde / (1 + dt * (f_el_prefactor * (2 * K_squared - KY**2)/K))

    return u1_tilde_new

def linear_first_order_scheme_for_u2(u2_tilde, dt, K_squared, K, f_el_prefactor, KX):

    u2_tilde_new = -u2_tilde / (1 + dt * (f_el_prefactor * (2 * K_squared - KX**2)/K))

    return u2_tilde_new
```

## Non linear

```python
def fourier_derivative_x(field_tilde, KX): ## returns in fourier space
    return 1j * KX * field_tilde

def fourier_derivative_y(field_tilde, KY): ## returns in fourier space
    return 1j * KY * field_tilde

def derivative_x(field_tilde, KX): ## returns in real space

    dfield_dx_tilde = fourier_derivative_x(field_tilde, KX)  # Differentiate in Fourier space
    dfield_dx = ifft_transform(dfield_dx_tilde)  # Transform back to real space
    
    return dfield_dx

def derivative_y(field_tilde, KY): ## returns in real space

    dfield_dy_tilde = fourier_derivative_y(field_tilde, KY)  # Differentiate in Fourier space
    dfield_dy = ifft_transform(dfield_dy_tilde)  # Transform back to real space
    
    return dfield_dy

def compute_gradient(field_tilde, KX, KY): ## returns in real space

    dfield_dx = derivative_x(field_tilde, KX)
    dfield_dy = derivative_y(field_tilde, KY)
    
    grad_field = np.stack([dfield_dx, dfield_dy], axis=0)  # Store as vector field
    return grad_field

def cross_derivative(field_tilde, KX, KY): ## returns in real space

    dfield_dy_tilde = fourier_derivative_y(field_tilde, KY)
    d2field_dy_dx = derivative_x(dfield_dy_tilde, KX)

    return d2field_dy_dx


def laplacian(field_tilde, K_squared): ## returns in real space

    laplacian_tilde = -K_squared * field_tilde  # Laplacian in Fourier space
    laplacian = ifft_transform(laplacian_tilde)  # Transform back to real space

    return laplacian

def compute_hessian_diagonal(w_tilde, KX, KY): ## returns in real space

    d2field_dx2_tilde = -(KX**2) * w_tilde  # Compute second derivative in x
    d2field_dy2_tilde = -(KY**2) * w_tilde  # Compute second derivative in y
    
    d2field_dx2 = ifft_transform(d2field_dx2_tilde)  # Transform back to real space
    d2field_dy2 = ifft_transform(d2field_dy2_tilde)  # Transform back to real space
    
    hessian_diag = np.stack([d2field_dx2, d2field_dy2], axis=0)  # Store as vector field
    return hessian_diag
```

## Non linear W

```python
def non_linear_1st_term(w_tilde, u1_tilde, u2_tilde, KX, KY): ## returns in real space

    d2w_dxy = cross_derivative(w_tilde, KX, KY)

    hessian_w = compute_hessian_diagonal(w_tilde, KX, KY)

    d2w_dx2 = hessian_w[0]
    d2w_dy2 = hessian_w[1]

    du1_dx = derivative_x(u1_tilde, KX)
    du2_dy = derivative_y(u2_tilde, KY)

    term1 = 2 * d2w_dx2 * du1_dx
    term2 = 2 * d2w_dxy * (du1_dx + du2_dy)
    term3 = 2 * d2w_dy2 * du2_dy

    total_term = term1 + term2 + term3

    return total_term

def non_linear_2nd_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared): ## returns in real space

    grad_w = compute_gradient(w_tilde, KX, KY)

    grad2_u_1 = laplacian(u1_tilde, K_squared)
    grad2_u_2 = laplacian(u2_tilde, K_squared)

    grad2_u = np.stack([grad2_u_1, grad2_u_2], axis=0)

    total_term = np.sum(grad_w * grad2_u, axis=0)  # Element-wise dot product

    return total_term

def non_linear_3rd_term(w_tilde, u1_tilde, u2_tilde, KX, KY): ## returns in real space

    grad_w = compute_gradient(w_tilde, KX, KY)

    du1_dx = derivative_x(u1_tilde, KX)
    du2_dy = derivative_y(u2_tilde, KY)

    div_u = du1_dx + du2_dy

    grad_div_u = compute_gradient(fft_transform(div_u), KX, KY)

    total_term = np.sum(grad_w * grad_div_u, axis=0)  # Element-wise dot product

    return total_term

def non_linear_4th_term(w_tilde, p, non_linear_prefactor, K_squared): ## returns in real space

    laplacian_w = laplacian(w_tilde, K_squared)

    w = ifft_transform(w_tilde)

    w_term = w**2 + (p/non_linear_prefactor)*w

    total_term = laplacian_w * w_term

    return total_term

def non_linear_5th_term(w_tilde, u1_tilde, u2_tilde, KX, KY): ## returns in real space

    grad_w = compute_gradient(w_tilde, KX, KY)

    mod_sq_w = np.sum(grad_w**2, axis=0)

    grad_mod_sq_w = compute_gradient(fft_transform(mod_sq_w), KX, KY)
    
    total_term = np.sum(grad_w * grad_mod_sq_w, axis=0)  # Element-wise dot product

    return total_term
```

```python
def non_linear_first_order_scheme_for_w(w_tilde, u1_tilde, u2_tilde, dt, K_fourth, K_squared, K, p, non_linear_prefactor, gamma0):

    first_term = non_linear_1st_term(w_tilde, u1_tilde, u2_tilde, KX, KY)
    second_term = non_linear_2nd_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared)
    third_term = non_linear_3rd_term(w_tilde, u1_tilde, u2_tilde, KX, KY)
    fourth_term = non_linear_4th_term(w_tilde, p, non_linear_prefactor, K_squared) # blows up
    fifth_term = non_linear_5th_term(w_tilde, u1_tilde, u2_tilde, KX, KY) # slower blow up

    total_term = first_term + second_term + third_term# + fourth_term + fifth_term

    w_tilde_new = dt * total_term/ (1 + dt * (K_fourth - gamma0 * K_squared + 2*K*f_el_prefactor))
    
    return w_tilde_new
```

## Non linear U

```python
def f_el_matrix_part(u1_tilde, u2_tilde, KX, KY, K, f_el_prefactor):
    
    M12 = (KX * KY) / K
    M21 = (KX * KY) / K
    
    f_el_u1 =  M12 * u2_tilde
    f_el_u2 = M21 * u1_tilde 

    f_el_u1_real = f_el_prefactor * ifft_transform(f_el_u1)
    f_el_u2_real = f_el_prefactor * ifft_transform(f_el_u2)
    
    return np.stack([f_el_u1_real, f_el_u2_real], axis=0)

def u_non_linear_1st_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared): ## returns in real space

    lap_u1 = laplacian(u1_tilde, K_squared)
    lap_u2 = laplacian(u2_tilde, K_squared)

    return np.stack([lap_u1, lap_u2], axis=0)

def u_non_linear_2nd_term(w_tilde, u1_tilde, u2_tilde, KX, KY): ## returns in real space

    du1_dx = derivative_x(u1_tilde, KX)
    du2_dy = derivative_y(u2_tilde, KY)

    div_u = du1_dx + du2_dy

    grad_div_u = compute_gradient(fft_transform(div_u), KX, KY)

    return grad_div_u

def u_non_linear_3rd_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared): ## returns in real space

    lap_w = laplacian(w_tilde, K_squared)
    grad_w = compute_gradient(w_tilde, KX, KY)

    return lap_w * grad_w

def u_non_linear_4th_term(w_tilde, u1_tilde, u2_tilde, KX, KY): ## returns in real space

    grad_w = compute_gradient(w_tilde, KX, KY)

    hess_w = compute_hessian_diagonal(w_tilde, KX, KY)
    d2w_dxdy = cross_derivative(w_tilde, KX, KY)

    W11 = hess_w[0]
    W12 = d2w_dxdy
    W21 = d2w_dxdy
    W22 = hess_w[1]

    term1 = W11 * grad_w[0] + W12 * grad_w[1]
    term2 = W21 * grad_w[0] + W22 * grad_w[1]

    return np.stack([term1, term2], axis=0)
```

```python
def non_linear_first_order_scheme_for_u(w_tilde, u1_tilde, u2_tilde, dt, KX, KY, K_squared, K, fel_linear_prefactor, gamma0):

    first_term = u_non_linear_1st_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared)
    second_term = u_non_linear_2nd_term(w_tilde, u1_tilde, u2_tilde, KX, KY)
    third_term = u_non_linear_3rd_term(w_tilde, u1_tilde, u2_tilde, KX, KY, K_squared)
    fourth_term = u_non_linear_4th_term(w_tilde, u1_tilde, u2_tilde, KX, KY)

    total_term = first_term + second_term + third_term + fourth_term

    u1_tilde_new = dt * total_term[0] / (1 + dt * (f_el_prefactor * (2 * K_squared - KY**2)/K))
    u2_tilde_new = dt * total_term[1] / (1 + dt * (f_el_prefactor * (2 * K_squared - KX**2)/K))

    return u1_tilde_new, u2_tilde_new
```
